```python
import pandas as pd
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
# Combine anomaly scores for each value of r


for r in range(1, 10):
    english_scores = pd.read_csv(f'r{r}_english.txt', header=None, names=['Score'])
    tagalog_scores = pd.read_csv(f'r{r}_tagalog.txt', header=None, names=['Score'])

    # Add labels (0 for English, 1 for Tagalog)
    english_scores['Label'] = 0
    tagalog_scores['Label'] = 1

    # Concatenate scores
    combined_scores = pd.DataFrame()
    combined_scores = pd.concat([combined_scores, english_scores, tagalog_scores], ignore_index=True)

    # Read combined anomaly scores
    data = combined_scores

    # Compute ROC curve
    fpr, tpr, thresholds = roc_curve(data["Label"], data["Score"])

    # Compute AUC
    auc_value = roc_auc_score(data["Label"], data["Score"])
    print(f"AUC R{r}:", auc_value)

    # Plot ROC curve
    plt.figure(figsize=(4, 6))
    plt.plot(fpr, tpr, label=f'AUC={auc_value:.2f}')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve {r}')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()
    print(auc_value)
    print(len(fpr), len(tpr))

# Sort the combined scores by anomaly score
combined_scores = combined_scores.sort_values(by='Score')
```

```python
import os
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Check if two files are provided as arguments
import sys

def read_data_from_files(train_output_filename, test_output_filename, labels_output_filename):
    # Read the contents of train_output_filename
    with open(train_output_filename, 'r') as train_file:
        train_sequences = train_file.readlines()

    # Read the contents of test_output_filename
    with open(test_output_filename, 'r') as test_file:
        test_sequences = test_file.readlines()

    # Read the contents of labels_output_filename
    with open(labels_output_filename, 'r') as labels_file:
        test_labels = [int(line.strip()) for line in labels_file.readlines()]

    return train_sequences, test_sequences, test_labels
    
class NegativeSelectionClassifier:
    def __init__(self, n, r):
        self.n = n
        self.r = r
        self.detectors = set()

    def train(self, train_sequences):
        self.detectors = self.generate_detectors(train_sequences)

    def generate_detectors(self, sequences):
        detectors = set()
        for sequence in sequences:
            for i in range(len(sequence) - self.n + 1):
                pattern = sequence[i:i+self.n]
                for j in range(len(pattern) - self.r + 1):
                    subpattern = pattern[j:j+self.r]
                    detectors.add(subpattern)
        return detectors

    def classify(self, test_sequences):
        anomaly_scores = []
        for sequence in test_sequences:
            score = self.compute_score(sequence)
            anomaly_scores.append(score)
        return anomaly_scores

    def compute_score(self, sequence):
        score = 0
        for i in range(len(sequence) - self.n + 1):
            pattern = sequence[i:i+self.n]
            found = False
            for j in range(len(pattern) - self.r + 1):
                subpattern = pattern[j:j+self.r]
                if subpattern in self.detectors:
                    found = True
                    break
            if not found:
                score += 1
        score /= len(sequence)  # Normalize by the length of the sequence
        return score

def merge_lists(list1, list2):
    merged_list = []
    index = 0

    for num_elements in list1:
        sublist = list2[index:index+num_elements]  # Extract sublist from list2
        sublist_average = sum(sublist) / len(sublist)  # Compute average of sublist
        merged_list.append(sublist_average)
        index += num_elements

    return merged_list
        


file_paths = ["syscalls/snd-unm/snd-unm.", "syscalls/snd-cert/snd-cert."]
file_numbers = ["1", "2", "3"]

for file_path in file_paths:
    for file_number in file_numbers:

        test_filename = file_path + file_number + ".test"
        labels_filename = file_path + file_number + ".labels"
        train_filename = file_path + "train"
        
        # Check if the test file exists
        if not os.path.isfile(test_filename):
            print("Data file {} not found.".format(test_filename))
            sys.exit(1)
        
        # Check if the labels file exists
        if not os.path.isfile(labels_filename):
            print("Labels file {} not found.".format(labels_filename))
            sys.exit(1)
        
        # Check if the train file exists
        if not os.path.isfile(train_filename):
            print("train file {} not found.".format(train_filename))
            sys.exit(1)
        # Create the output filenames
        test_output_filename = test_filename + ".split"
        #labels_output_filename = labels_filename + ".split"
        labels_output_filename = labels_filename
        train_output_filename = train_filename + ".split"
        
        # Check if the output files exist and delete them if they do
        if os.path.isfile(test_output_filename):
            os.remove(test_output_filename)
        
        if os.path.isfile(labels_output_filename):
            l = 0;
            #os.remove(labels_output_filename)
        
        if os.path.isfile(train_output_filename):
            os.remove(train_output_filename)
            
        # Set the minimum length
        min_length = 40

        chunk_index = []
        # Read both data and labels files simultaneously
        with open(test_filename, 'r') as test_file:
            with open(test_output_filename, 'w') as test_output:
                for test_line in test_file:
                    i = 0
                    test_line = test_line.strip()
               
                    # Split the line into multiple lines with a maximum length of min_length
                    while test_line:
                        i+=1
                    #while len(test_line) >= min_length:
                        # Write the current portion of the line to the output files
                        test_output.write(test_line[:min_length] + '\n')
                        test_line = test_line[min_length:]
                    chunk_index.append(i)
                    
        
        with open(train_filename, 'r') as train_file:
            with open(train_output_filename, 'w') as train_output:
                for train_line in train_file:
                    train_line = train_line.strip()
                    # Split the line into multiple lines with a maximum length of min_length
                    while train_line:
                    #while len(test_line) >= min_length:
                        # Write the current portion of the line to the output files
                        train_output.write(train_line[:min_length] + '\n')
                        train_line = train_line[min_length:]
        
        print("Output written to {} and {} and {}".format(test_output_filename, labels_output_filename, train_output_filename))
        r_values = [2, 5, 7, 10, 15, 20]
        n_values = [2, 5, 7, 10, 15, 20]
        #train_sequences, test_sequences, test_labels = read_data_from_files(train_output_filename, test_output_filename, labels_output_filename)
        train_sequences, test_sequences, test_labels = read_data_from_files(train_output_filename, test_output_filename, labels_filename)

        #Example data
        #train_sequences = ["abcb", "defg", "hijk"]
        #test_sequences = ["abcb", "defg", "hijj"]
        #test_labels = [0, 0, 1]
        
        #r_values = [4]
        #n_values = [4]
        
        print(len(test_labels), len(test_sequences), len(train_sequences))
        print(train_output_filename, test_output_filename, labels_output_filename)
        highest_auc_score = 0
        best_r = None
        best_n = None
        best_fpr = None
        best_tpr = None
        
        for r in r_values:
            for n in n_values:
                if r > n:
                    continue
                
                # Initialize and train the classifier
                classifier = NegativeSelectionClassifier(n, r)
                classifier.train(train_sequences)
                
                # Classify test sequences
                anomaly_scores = classifier.classify(test_sequences)
                
                # Evaluate classification performance
                auc = roc_auc_score(test_labels, merge_lists(chunk_index, anomaly_scores))
                print("r: {}, n: {}, AUC Score: {}".format(r, n, auc))
                
                # Check if the current AUC score is the highest
                if auc > highest_auc_score:
                    highest_auc_score = auc
                    best_r = r
                    best_n = n
                    
                    # Compute ROC curve
                    fpr, tpr, thresholds = roc_curve(test_labels, merge_lists(chunk_index, anomaly_scores))
                    best_fpr = fpr
                    best_tpr = tpr
                        
        # Plot ROC curve for the best AUC score
        plt.figure(figsize=(4, 6))
        plt.plot(best_fpr, best_tpr, label=f'Best AUC={highest_auc_score:.2f}')
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve (Best AUC)')
        plt.legend(loc='lower right')
        plt.grid(True)
        plt.show()
        
        print(f"Highest AUC Score: {highest_auc_score}, r: {best_r}, n: {best_n}")
                
```
